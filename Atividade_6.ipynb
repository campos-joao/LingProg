1. Defina a função soma_nat que recebe como argumento um número natural n e devolve a soma de todos os números naturais até n.


```python
def soma_nat(numero): 
    if numero < 0: return 'O número não é natural'
    elif numero <= 1: return 1
    
    else: return numero + soma_nat(numero - 1) 
    
print(soma_nat(2))
print(soma_nat(1))
print(soma_nat(0))
print(soma_nat(10))
print(soma_nat(-10))
```

2 Defina a função div que recebe como argumentos dois números naturais m e n e devolve o resultado da divisão inteira de m por n. Neste exercício você não pode recorrer às operações aritméticas de multiplicação, divisão e resto da divisão inteira.

```python
def div(m,n):
    if m < n:
        return 0
    
    else:
        return 1 + div(m-n,n)
    
print (div(20,5))
```

Defina a função prod_lista que recebe como argumento uma lista de inteiros e devolve o produto dos seus elementos.


```python
def prod_lista(i):
        if len(i) == 1:
            return (i[0])
        
        elif len(i) == 2:
            return (i[0] * i[1])
        
        else:
            return (i[0] * prod_lista(i[1:]))
        
prod_lista([2,3,4,5,7,8])
```

Defina a função contem_parQ que recebe como argumento uma lista de números inteiros w e devolve True se w contém um número par e False em caso contrário.


```python
def contem_parQ(w):
    if len(w) == 0:
        return False
    
    else:
        a = w.pop()
        if a%2 == 0:
            return True
        
        else:
            return (contem_parQ(w))
        
print(contem_parQ([2,33,55,77,88]))
print(contem_parQ([1,3,5,7,9]))
```

Defina a função todos_imparesQ que recebe como argumento uma lista de números inteiros w e devolve True se w contém apenas números ímpares e False em caso contrário.


```python
def contem_imparesQ(w):
    if len(w) == 0:
        return True
    else:
        a = w.pop()
        if a%2 == 0:
            return False
        else:
            return contem_imparesQ(w)
        
print(contem_imparesQ([11,33,44,55,66,77,88]))
print(contem_imparesQ([1,3,5,7,9]))
```
